```python
import openpyxl
import pandas as pd
from openpyxl.styles import PatternFill, Font, Alignment
import random

#Open the Excel file
excel_file = 'C:\\Users\\AVALOSCO\\Desktop\\Visor cs.xlsx'

# Name of the sheet you want to read
SRT01 = 'SRT01'
# Number of the first row with data
first_row = 4

workbook = openpyxl.load_workbook(excel_file, data_only=True)

# Select the sheet
sheet = workbook[SRT01 ]

# Read data from were the Data Base starts
data = []
for row in sheet.iter_rows(min_row=first_row, max_col=12, values_only=True):
    data.append(row)

column_names = ['Seq', 'PO Date', 'PO Num', '1st', 'Model', 'PO Qty', 'Configuration', 'Diameter', 
                'Expander', 'Ecoated', 'Material', 'ReHeat']

df = pd.DataFrame(data, columns=column_names)
df['PO Date'] = df['PO Date'].astype(str).str[:11]

print(df.head(5))
```

```python
# Ordenar el DataFrame por "Configuration" y "Material"
df_sorted = df.sort_values(by=['PO Date', 'Configuration','Material', 'Diameter'])

# Inicializar una lista para almacenar los grupos de datos
grouped_data = []
current_group = []
consecutive_m_count = 0
max_consecutive_M = 15  

for _, row in df_sorted.iterrows():
    if row['ReHeat'] == 'M':
        consecutive_m_count += 1
        if consecutive_m_count > max_consecutive_M:
            grouped_data.append(current_group)
            current_group = []
            consecutive_m_count = 0
    else:
        consecutive_m_count = 0
    current_group.append(row)

if current_group:
    grouped_data.append(current_group)

# Ahora, 'grouped_data' contiene grupos de datos donde no hay m치s de 15 "M" consecutivas en "ReHeat"
all_data = []
for group in grouped_data:
    all_data.extend(group)

workbook = openpyxl.Workbook()
sheet = workbook.active

# Considerar cual fue el material utilizado en la ultima corrida para poder arrancar con el mismo rollo
last_material_value = None

# Iterar sobre los datos y agregarlos a la hoja de c치lculo con diferentes colores
for row_idx, row in enumerate(all_data, start=2):
   configurations = df['Configuration'].unique()
color_dict = {config: PatternFill(start_color="%06x" % random.randint(0, 0xFFFFFF),
                                  end_color="%06x" % random.randint(0, 0xFFFFFF),
                                  fill_type="solid")
              for config in configurations}

# Obtener el valor en la columna "Material"
material_value = row['Material']
if not material_value:
            material_value = last_material_value 
            if not material_value and group:
                material_value = group[0]['Material']

#Colores
for row_idx, row in enumerate(all_data, start=2):
    configuration_value = row['Configuration']
    fill_color = color_dict.get(configuration_value, PatternFill(fill_type="solid"))  # Obtener el color correspondiente
    for col_idx, value in enumerate(row, start=1):
        sheet.cell(row=row_idx, column=col_idx, value=value).fill = fill_color

for col_idx, col_name in enumerate(column_names, start=1):
    sheet.cell(row=1, column=col_idx, value=col_name).fill = PatternFill(start_color="FFFFFF", end_color="FFFFFF", fill_type="solid")
    sheet.cell(row=1, column=col_idx).font = Font(bold=True)
    sheet.cell(row=1, column=col_idx).alignment = Alignment(horizontal="center")

#Columnas    
column_names = ['Seq', 'PO Date', 'PO Num', '1st', 'Model', 'PO Qty', 'Configuration', 'Diameter',
                'Expander', 'Ecoated', 'Material', 'ReHeat']

for col_idx, col_name in enumerate(column_names, start=1):
    sheet.cell(row=1, column=col_idx, value=col_name)

# Guardar el archivo Excel con nuevo nombre
output_excel = 'VCS_SRT01_Python.xlsx'
workbook.save(output_excel)

print(f"Todos los grupos han sido exportados a la hoja de c치lculo '{output_excel}'.")
```
